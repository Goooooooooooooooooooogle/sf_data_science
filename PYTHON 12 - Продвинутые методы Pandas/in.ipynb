```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
```

```python
display(melb_df.info())
```

```python
melb_df['Date']=pd.to_datetime(melb_df['Date'], dayfirst=True)
```

```python
display(melb_df['Date'].dt.quarter.value_counts())
```

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car']
max_unique_count = 150
for col in melb_df.columns:
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude:
        melb_df[col] = melb_df[col].astype('category')
display(melb_df.info())
```

```python
date05 = pd.to_datetime('2017-05-01')
date09 = pd.to_datetime('2017-09-01')

ot_do = (date05 <= melb_df['Date']) & (melb_df['Date'] <= date09)

melb_df[ot_do].groupby('SellerG')['Price'].sum().sort_values(ascending=True)
```

```python
dates = pd.read_csv('data/dates.csv')
pd.DatetimeIndex(dates['date']).year.value_counts()
```

```python
ratings1 = pd.read_csv('data/ratings1.csv')
ratings2 = pd.read_csv('data/ratings2.csv')

ratings = pd.concat(
    [ratings1, ratings2],
    ignore_index=True
)
display(ratings)
```

```python
ratings = ratings.drop_duplicates(ignore_index=True)
print('Число строк в таблице ratings: ', ratings.shape[0])
```

```python
ratings_dates = pd.concat([ratings, dates], axis=1)
display(ratings_dates.tail(7))
```

```python
movies = pd.read_csv('data/movies.csv')
joined_false = ratings_dates.join(
    movies,
    rsuffix='_right',
    how='left'
)
display(joined_false)
```

```python
joined = ratings_dates.join(
    movies.set_index('movieId'),
    on='movieId',
    how='left'
)
display(joined.head())
```

```python
merged = ratings_dates.merge(
    movies,
    on='movieId',
    how='left'
)
display(merged.head())
```

```python
print('Число строк в таблице ratings_dates: ', ratings_dates.shape[0])
print('Число строк в таблице merged: ', merged.shape[0])
print(ratings_dates.shape[0] == merged.shape[0])
```

```python
merged2 = ratings_dates.merge(
    movies,
    on='movieId',
    how='outer'
)
print('Число строк в таблице merged2: ', merged2.shape[0])
display(merged2.tail())
```
