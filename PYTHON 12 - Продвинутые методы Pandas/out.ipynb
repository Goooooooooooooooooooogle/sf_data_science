```python
import pandas as pd

ratings_movies = pd.read_csv('data/ratings_movies.csv')
```

```python
import re 
def get_year_release(arg):
    #находим все слова по шаблону "(DDDD)"
    candidates = re.findall(r'\(\d{4}\)', arg) 
    # проверяем число вхождений
    if len(candidates) > 0:
        #если число вхождений больше 0,
	#очищаем строку от знаков "(" и ")"
        year = candidates[0].replace('(', '')
        year = year.replace(')', '')
        return int(year)
    else:
        #если год не указан, возвращаем None
        return None
```

```python
ratings['date'] = pd.to_datetime(ratings['date'])
ratings.info()
```

```python
ratings['year_release'] = ratings['title'].apply(get_year_release)
ratings = ratings.drop(['get_year_release'], axis=1)
ratings.info()
```

```python
ratings = ratings_movies.copy()

ratings['year_release'] = ratings['title'].apply(get_year_release)

ratings1999=ratings[ratings['year_release'] == 1999]
 
ratings1999.groupby('title')['rating'].mean().sort_values()
```

```python
ratings2010=ratings[ratings['year_release'] == 2010]
 
ratings2010.groupby('genres')['rating'].mean().sort_values()
```
